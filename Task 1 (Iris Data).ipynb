## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sns.set_style('whitegrid')
```

## Load Dataset

```python
df = pd.read_csv('iris_dataset.csv')

df.head()
```

## Display Dataset Structure

```python
print("Dataset Shape:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nFirst Five Rows:")
print(df.head())
```

## Dataset Information

```python
df.info()
```

## Statistical Summary

```python
df.describe()
```

## Check Missing Values

```python
df.isnull().sum()
```

## Unique Species

```python
df['target'].value_counts()
```

## Scatter Plot (Sepal Length vs Sepal Width)

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x='sepal length (cm)',
    y='sepal width (cm)',
    hue='target',
    s=100
)

plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

## Scatter Plot (Petal Length vs Petal Width)

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x='petal length (cm)',
    y='petal width (cm)',
    hue='target',
    s=100
)

plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()
```

## Histogram of Sepal Length

```python
plt.figure(figsize=(8,6))

sns.histplot(
    data=df,
    x='sepal length (cm)',
    bins=15,
    kde=True
)

plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')

plt.show()
```

## Histograms of All Numerical Features

```python
df.hist(figsize=(12,8))

plt.tight_layout()
plt.show()
```

## Box Plot of Petal Length by Species

```python
plt.figure(figsize=(8,6))

sns.boxplot(
    data=df,
    x='target',
    y='petal length (cm)'
)

plt.title('Petal Length Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')

plt.show()
```

## Box Plot for All Numerical Features

```python
plt.figure(figsize=(10,6))

sns.boxplot(
    data=df.drop('target', axis=1)
)

plt.title('Box Plot of Numerical Features')

plt.show()
```

## Correlation Heatmap (Extra Professional Visualization)

```python
plt.figure(figsize=(8,6))

correlation = df.drop('target', axis=1).corr()

sns.heatmap(
    correlation,
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')

plt.show()
```

## Pair Plot (Optional but Highly Recommended)

```python
sns.pairplot(
    df,
    hue='target'
)

plt.show()
```
